```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#loading the input data
IN = np.loadtxt('IN.txt')

# cheking that it is normalized
print(np.sum(IN))

# plotting the data
plt.plot(IN)
```

```python
#loading the output data
OUT = np.loadtxt('OUT.txt')

# cheking that it is normalized
print(np.sum(OUT))

# plotting the data
plt.plot(OUT)
```

```python
#loading the transfer matrix
TR = np.loadtxt('TR.txt')

# cheking that it is normalized
print(np.sum(TR, axis = 0))

# plotting the data
plt.matshow(TR);
plt.colorbar()
plt.show()
```
