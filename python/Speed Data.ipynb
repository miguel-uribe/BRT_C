Analizing the speed profile

```python
import numpy as np
import pandas as pd
```

```python
data_0 = np.loadtxt("../0_2_4_6_8.txt")
```

```python
data_0.shape
```

```python
dataDF_0 = pd.DataFrame(data_0, columns = ['position', 'speed'])
```

```python
speedprofile_0=dataDF_0.groupby('position').mean()
```

```python
speedprofile_0.plot()
```
